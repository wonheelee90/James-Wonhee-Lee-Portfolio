```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
## Load data. Here are the columns.
## y: number of riders per day, and our target variable
## PRCP: precipitation
## SNOW: amount of snow
## TAVG: average temperature for the day
ts_data = pd.read_csv('data/green_taxi.csv')
ts_data = ts_data.set_index('ds')
ts_data.head()
```

```python
## We will generate predictions for the last 2 weeks of 2016, so we will create train/test sets based off of the date.
train_end_date = '2016-12-17'
train = ts_data[ts_data.index <= train_end_date]
test = ts_data[ts_data.index > train_end_date]
train.shape, test.shape
```

```python
## Initiate the VAR model, set initial order to 7. We will let the model fit identify the best set of lags using the AIC metric
model = VAR(endog=train)
model.select_order(7)
model_fit = model.fit(maxlags=7, ic='aic')
model_fit.summary()
```

```python
## Now that we've trained the model, we can proceed to make forecasts.
prediction = model_fit.forecast(model_fit.y, steps=len(test))
prediction
```

```python
## Store predictions in dataframe
yhat = pd.DataFrame(prediction, index = test.index)[0]
y = test['y']
results = pd.concat([yhat, y], axis=1)
results.columns = ['preds', 'y']
results
```

```python
## Let's do some quick error calculation, mean absolute error and mean squared error
mae = metrics.mean_absolute_error(results['y'], results['preds'])
mse = metrics.mean_squared_error(results['y'], results['preds'])
mae, mse
```

```python
## Visualize predictions vs. actual
plt.figure(figsize=(20,8))
plt.plot(results['y'], label = 'actual')
plt.plot(results['preds'], label = 'predicted')
plt.legend()
plt.show()
## It seems like the predictions tend to be off for end-of-year holidays; it would make sense to add a "holiday" feature for the next step!
```
